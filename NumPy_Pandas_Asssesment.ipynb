# Employee Salary Adjustment

A company stores salaries in a NumPy array.
salaries = np.array([45000, 52000, 61000, 48000, 75000])
Tasks:
1. Increase all salaries by 10%.
2. Find the average salary.
3. Find the maximum and minimum salary.
4. Print salaries greater than 60000.


```python
salaries= np.array([45000,52000,61000,48000,75000])
```

```python
# increasing all salaries by 10%

new_salaries= salaries*1.10
new_salaries
```

```python
# findingg average salary

avg_salary_old= np.average(salaries)
avg_salary_new= np.average(new_salaries)

print("Average Salary Before Increase: ", avg_salary_old)
print("Average Salary After Increase: ", avg_salary_new)
```

```python
# finding the Maximum and Minimum Salary

max_salary_old= np.max(salaries)
min_salary_old= np.min(salaries)

max_salary_new= np.max(new_salaries)
min_salary_new= np.min(new_salaries)

print("Maximum Salary Before Increase: ", max_salary_old)
print("Minimum Salary Before Increase: ", min_salary_old)
print("Maximum Salary After Increase: ", max_salary_new)
print("Minimum Salary After Increase: ", min_salary_new)
```

```python
# finding salaries greater than 60000

salaries_greater_than_60000= salaries[salaries>60000]
salaries_greater_than_60000
```

# Sales Data Analysis

Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])
Tasks:
1. Compute total sales.
2. Compute mean and standard deviation.
3. Identify days where sales exceeded the average.
4. Normalize the sales using Min-Max scaling



```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])
```

```python
total_sales=np.sum(sales)
print(' Total_sales is ',total_sales)
```

```python
# Mean and standard deviation

mean_sales= np.mean(sales)
std_sales= np.std(sales)

print(' Mean_sales is ',mean_sales)
print(' Std_sales is', std_sales)
```

```python
# To identify days where sales exceeded the average

days_above_average_sales= sales[sales>mean_sales]
days_above_average_sales
```

```python
# Normalize the sales using Min-Max scaling.

min_sales= np.min(sales)
max_sales= np.max(sales)

normalized_sales= (sales-min_sales)/(max_sales-min_sales)
normalized_sales
```

# Missing Data Handling

Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
Tasks:
1. Detect missing values.
2. Replace missing Age with mean age.
3. Replace missing Salary with median salary.
4. Remove rows with any remaining missing values.

```python
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
```

```python
Data_df= pd.DataFrame(data)
Data_df
```

```python
# Missing Values

missing_values= Data_df.isnull()
missing_values
```

```python
# Replacing Age with missing values

mean_age = Data_df['Age'].mean()
Data_df['Age'].fillna(mean_age, inplace=True)
Data_df
```

```python
# Replace missing Salary with median

median_salary= Data_df['Salary'].median()
Data_df['Salary'].fillna(median_salary, inplace=True)

Data_df
```

```python
# Remove rows with any remaining missing values

Data_df.dropna(inplace=True)
Data_df
```

# Customer Purchase Analysis

Dataset:
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

asks:
1. Find the average purchase amount by city.
2. Identify customers with purchase above overall average.
3. Find the city generating the highest revenue.
4. Create a column Age_Group:
o <30 → Young
o 30–50 → Adult
o 50 → Senior
5. Find average purchase per age group.

```python
customer_data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
```

```python
customer_df = pd.DataFrame(customer_data)
customer_df
```

```python
# Average purchase amount by city

avg_by_city = customer_df.groupby('City')['PurchaseAmount'].mean()
print(avg_by_city)
```

```python
# Customers with purchase above overall average

overall_avg = customer_df['PurchaseAmount'].mean()

high_value_customers = customer_df[customer_df['PurchaseAmount'] > overall_avg]
print(high_value_customers)
```

```python
# City generating highest revenue

total_by_city = customer_df.groupby('City')['PurchaseAmount'].sum()
top_city = total_by_city.idxmax()

print(total_by_city)
print("Top city:", top_city)
```

```python
# Age Group Column

def age_group(age):
    if age < 30:
        return "Young"
    elif age <= 50:
        return "Adult"
    else:
        return "Senior"

customer_df['Age_Group'] = customer_df['Age'].apply(age_group)
customer_df
```

```python
# Average purchase per age group

avg_by_age_group = customer_df.groupby('Age_Group')['PurchaseAmount'].mean()
print(avg_by_age_group)
```

# Time Series Sales Analysis

Generate data:
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
Tasks:
1. Convert Date to datetime index.
2. Compute monthly percentage growth.
3. Calculate 3-month rolling average sales.
4. Identify the month with highest growth rate.

```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
```

```python
df
```

```python
# Convert Date to datetime index

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

print(df)
```

```python
# Monthly percentage growth

df['Monthly_Growth_%'] = df['Sales'].pct_change() * 100
print(df)
```

```python
# 3-month rolling average

df['Rolling_3M_Avg'] = df['Sales'].rolling(window=3).mean()
print(df)
```

```python
# Month with highest growth rate

max_growth_month = df['Monthly_Growth_%'].idxmax()
max_growth_value = df['Monthly_Growth_%'].max()

print(max_growth_month, max_growth_value)
```
